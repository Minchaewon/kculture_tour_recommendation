```python
import pandas as pd
import numpy as np

area = ['chungcheon', 'daegoo', 'daejeon', 'gangwon', 'gyeonggi', 'incheaon', 'jeju', 'jeonla', 'sejeon', 'seoul', 'lsan', 'busan','gyeongsang','gwangju']
change_area = ['충청', '대구', '대전', '강원', '경기', '인천', '제주', '전라', '세종', '서울', '울산', '부산', '경상', '광주']

merged_df = pd.DataFrame()

for i in range(len(area)):
    # CSV 파일 열기
    data = pd.read_csv(f"C:/Users/mcw08/OneDrive/바탕 화면/동덕여대/2025 문화 디지털혁신 및 데이터 활용 공모전/새로운 데이터/카카오맵 크롤링 데이터(맛집)/{area[i]}.csv", encoding='utf-8')
    # addr 열의 \n 제거
    data['Addr'] = data['Addr'].str.replace('\n', '')
    # ADDR_area 열 추가
    data['ADDR_area'] = change_area[i]
    merged_df = pd.concat([merged_df, data])
    
merged_df['ADDR_area'].unique()
```

```python
# score 3.5이상인 데이터만 보존 : 5,464건
count = merged_df[merged_df['Score'] >= 3.5].shape[0]
filtered_df = merged_df[merged_df['Score'] >= 3.5]
filtered_df
```

```python
filtered_df['Category'].unique()
```

```python
# Category에서 불필요한 열이 포함된 (ex. 산봉우리, 봉수대...)행, null 값 제거
filtered_df['Category'].unique()
filtered_df = filtered_df[~filtered_df['Category'].isin(['광장','섬(내륙)','하천','문','강','문화유적','호수','릉,묘,총','산','먹자골목','도보여행','사당,제단','관광안내소','동상','간식','둘레길','향교,서당','탑,비석','카지노','생가,고택','종교유적지','고개','자전거여행','등산로','산봉우리','봉수대','샘터,약수터','숲','바위','저수지'])]
filtered_df = filtered_df.dropna(subset=['Category'])
filtered_df['Category'].unique()
```

```python
# cafe, playground, restaurant, stay, staion, store, cvs, shop으로 카테고리 통합
merged_cate = filtered_df
categories = ['제과,베이커리','디저트카페','카페','커피전문점','떡,한과','갤러리카페','다방','도넛','샌드위치','전통찻집','북카페','아이스크림','카페거리']
replace_cate = 'cafe'
merged_cate['Category'] = filtered_df['Category'].replace(categories, replace_cate)

merged_cate = filtered_df
categories = ['도자기,도예촌','천문대','실내동물원','유적지','자연휴양림','산성,성곽','온천','계곡','동물원','폭포','전망대','테마거리','수목원,식물원','고궁,궁','테마파크','관광,명소']
replace_cate = 'playground'
merged_cate['Category'] = filtered_df['Category'].replace(categories, replace_cate)
merged_cate['Category'].unique()

excluded_categories = ['cafe','playground']
replace_cate = 'restaurant'
mask = ~merged_cate['Category'].isin(excluded_categories)
merged_cate.loc[mask,'Category'] = replace_cate
merged_cate['Category'].unique()
merged_cate
```
