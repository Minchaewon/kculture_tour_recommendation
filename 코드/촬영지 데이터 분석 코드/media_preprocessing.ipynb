* 배우이름 포함 허용 조건
1) 한글 2~4글자 : 성 + 이름
2) 성+이름/이름 구조 : ex) 신혜정/혜정
3) 영문+한글 조합 : ex) EXID 하니

```python
import re

# 허용 조건 정의
def is_valid_actor_name(name):
    return (
        re.fullmatch(r"[가-힣]{2,4}", name) or                         # 조건 1: 한글 2~4글자
        re.fullmatch(r"[가-힣]{3}/[가-힣]{2,3}", name) or              # 조건 2: 성+이름/이름
        re.fullmatch(r"[A-Z]+ ?[가-힣]{2,3}", name)                   # 조건 3: 영문 + 한글 이름
    )

# 리스트에 조건 위배된 항목이 하나라도 있는지 확인
def has_invalid_actor(name_list):
    return any(not is_valid_actor_name(name) for name in name_list)

# 조건 위배된 행 추출
invalid_actor_rows = media_drama[media_drama['배우이름'].apply(has_invalid_actor)]

invalid_actor_rows
```

```python
# 위 조건에 맞지 않는 항목은 리스트에서 제거
def remove_invalid_actor_names(name_list):
    return [name for name in name_list if is_valid_actor_name(name)]

# 적용
media_drama['배우이름'] = media_drama['배우이름'].apply(remove_invalid_actor_names)
```

```python
import requests

API_KEY = "P01QIXV046V95475540D"
BASE_URL = "http://api.koreafilm.or.kr/openapi-data2/wisenut/search_api/search_json2.jsp"

def get_actors_from_kmdb(title):
    params = {
        "ServiceKey": API_KEY,
        "collection": "kmdb_new2",  # ✅ 꼭 수정
        "query": title,
        "detail": "Y"
    }

    try:
        response = requests.get(BASE_URL, params=params)
        response.encoding = 'utf-8'
        data = response.json()
    except Exception as e:
        print(f"[❌ JSON 오류] {title}: {e}")
        print(response.text[:300])
        return []

    actor_names = []
    try:
        for item in data['Data'][0]['Result']:
            actors = item.get("actors", {}).get("actor", [])
            actor_names += [actor["actorNm"] for actor in actors]
    except Exception as e:
        print(f"[❌ 데이터 파싱 오류] {title}: {e}")
    
    return list(set(actor_names))
```

```python
# 1. 크롤링: 네이버에서 그룹 멤버 자동 수집
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
import re
from collections import defaultdict

# --- STEP 1: 그룹 멤버 크롤링 함수 ---
def get_naver_group_members(group_name):
    headers = {'User-Agent': 'Mozilla/5.0'}
    query = f"{group_name} 멤버"
    url = f"https://search.naver.com/search.naver?query={query}"
    
    res = requests.get(url, headers=headers)
    if res.status_code != 200:
        return []
    
    soup = BeautifulSoup(res.text, 'html.parser')
    members = []

    selector1 = '#main_pack section.sc_new.cs_common_module.case_empasis._au_people_content_wrap._people_star div.scroll_img_vertical_75_84.type_member ul li a div.title_box strong'
    tags = soup.select(selector1)
    
    if tags:
        for tag in tags:
            name = tag.get_text(strip=True)
            if name:
                members.append(name)
    else:
        selector2 = '#main_pack section.sc_new.cs_common_module.case_normal._au_people_content_wrap div.cm_info_box.type_anniversary div.detail_info._detail_info_wrap dl div dd span a'
        tags = soup.select(selector2)
        for tag in tags:
            name = tag.get_text(strip=True)
            if name:
                members.append(name)

    if not members:
        members.append(group_name)
        
    return members

# --- STEP 2: 그룹 목록 불러와서 멤버 수집 ---
group_list = media_artist["제목"].unique().tolist()
group_members_dict = {}

for group in group_list:
    try:
        members = get_naver_group_members(group)
        print(f"{group} 멤버 크롤링 완료: {members}")
        group_members_dict[group] = members
        time.sleep(1)
    except Exception as e:
        print(f"{group} 멤버 크롤링 실패:", e)
```

```python
# 2. 멤버명 보완 (성을 뺀 이름 추가)
group_members_dict_extended = defaultdict(list)

for group, members in group_members_dict.items():
    extended_members = []
    for name in members:
        name = name.strip()
        extended_members.append(name)
        if len(name) >= 3:  # '정진운'처럼 3글자 이상이면
            short_name = name[1:]  # '진운'
            if len(short_name) >= 2:  # 1글자는 제외 ('이', '비' 방지)
                extended_members.append(short_name)
    group_members_dict_extended[group] = list(dict.fromkeys(extended_members))  # 중복 제거
group_members_dict_extended
```

```python
import re
from collections import defaultdict

# 1. 성을 뺀 이름도 포함한 확장 멤버 dict 생성
group_members_dict_extended = defaultdict(list)

for group, members in group_members_dict.items():
    extended_members = []
    for name in members:
        name = name.strip()
        extended_members.append(name)
        if len(name) >= 3:
            short_name = name[1:]
            if len(short_name) >= 2:
                extended_members.append(short_name)
    group_members_dict_extended[group] = list(dict.fromkeys(extended_members))

# 2. 전체 아티스트명 (그룹명 + 멤버명)
all_names = []
for names in group_members_dict_extended.values():
    all_names.extend(names)

all_artist_names = sorted(set(list(group_members_dict_extended.keys()) + all_names), key=len, reverse=True)
name_pattern = re.compile('|'.join(map(re.escape, all_artist_names)))

# 3. 해당 그룹 멤버만 정규식 생성
def get_group_pattern(group):
    members = group_members_dict_extended.get(group, [])
    candidates = sorted([group] + members, key=len, reverse=True)
    return re.compile('|'.join(map(re.escape, candidates)))

# 4. 장소설명에서 그룹/타 그룹 아티스트 분리 추출
def extract_group_and_other_artists(row):
    text = row['장소설명']
    if pd.isna(text):
        return pd.Series([None, None])

    group = row['제목']
    group_pattern = get_group_pattern(group)

    # 1) 해당 그룹 아티스트 추출
    group_found = group_pattern.findall(text)
    group_found = list(dict.fromkeys(group_found))

    # 2) 전체에서 아티스트 추출 후, 그룹에서 찾은 건 제외
    all_found = name_pattern.findall(text)
    all_found = list(dict.fromkeys(all_found))
    other_found = [name for name in all_found if name not in group_found]

    return pd.Series([
        group_found if group_found else None,
        other_found if other_found else None
    ])

# 5. 적용
media_artist[['아티스트_1차', '아티스트_기타']] = media_artist.apply(extract_group_and_other_artists, axis=1)
media_artist
```
